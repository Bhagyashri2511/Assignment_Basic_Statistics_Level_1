# Q12)  Below are the scores obtained by a student in tests 
34,36,36,38,38,39,39,40,40,41,41,41,41,42,42,45,49,56
1)	Find mean, median, variance, standard deviation.
2)	What can we say about the student marks? 

```python
scores = [34, 36, 36, 38, 38, 39, 39, 40, 40, 41, 41, 41, 41, 42, 42, 45, 49, 56]
```

```python
scores
```

```python
mean = sum(scores) / len(scores)
```

```python
print("Mean:", mean)
```

```python
sorted_scores = sorted(scores)
n = len(sorted_scores)
```

```python
median = sorted_scores[n // 2] if n % 2 != 0 else (sorted_scores[(n // 2) - 1] + sorted_scores[n // 2]) / 2
```

```python
print("Median:", median)
```

```python
squared_deviations = [(x - mean)**2 for x in scores]
```

```python
variance = sum(squared_deviations) / (len(scores) - 1)
```

```python
print("Variance:", variance)
```

```python
import math
```

```python
std_deviation = math.sqrt(variance)
print("Standard Deviation:", std_deviation)
```

# Q 21) Check whether the data follows normal distribution
a)	Check whether the MPG of Cars follows Normal Distribution 
        Dataset: Cars.csv

```python
plt.figure(figsize=(10, 6))
plt.hist(MPG, bins=20, edgecolor='k')
plt.title('Histogram of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
stats.probplot(MPG, plot=plt)
plt.title('Q-Q Plot of MPG')
plt.show()
```

```python
statistic, p_value = stats.shapiro(MPG)
alpha = 0.05
print("Shapiro-Wilk Test:")
print("Test Statistic:", statistic)
print("p-value:", p_value)
if p_value > alpha:
    print("Data follows a normal distribution (p > 0.05)")
else:
    print("Data does not follow a normal distribution (p <= 0.05)")
```

# b)	Check Whether the Adipose Tissue (AT) and Waist Circumference(Waist)  from wc-at data set  follows Normal Distribution 
       Dataset: wc-at.csv

```python
data = pd.read_csv('wc-at.csv')
```

```python
AT = data['AT']
Waist = data['Waist']
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(AT, bins=20, edgecolor='k')
plt.title('Histogram of Adipose Tissue (AT)')
plt.xlabel('AT')
plt.ylabel('Frequency')
```

```python
plt.subplot(1, 2, 2)
plt.hist(Waist, bins=20, edgecolor='k')
plt.title('Histogram of Waist Circumference (Waist)')
plt.xlabel('Waist')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
stats.probplot(AT, plot=plt)
plt.title('Q-Q Plot of Adipose Tissue (AT)')
plt.subplot(1, 2, 2)
stats.probplot(Waist, plot=plt)
plt.title('Q-Q Plot of Waist Circumference (Waist)')
plt.tight_layout()
plt.show()
```

```python
statistic_AT, p_value_AT = stats.shapiro(AT)
alpha = 0.05
print("Shapiro-Wilk Test for Adipose Tissue (AT):")
print("Test Statistic:", statistic_AT)
print("p-value:", p_value_AT)
if p_value_AT > alpha:
    print("AT follows a normal distribution (p > 0.05)")
else:
    print("AT does not follow a normal distribution (p <= 0.05)")
```

```python
statistic_Waist, p_value_Waist = stats.shapiro(Waist)
print("Shapiro-Wilk Test for Waist Circumference (Waist):")
print("Test Statistic:", statistic_Waist)
print("p-value:", p_value_Waist)
if p_value_Waist > alpha:
    print("Waist follows a normal distribution (p > 0.05)")
else:
    print("Waist does not follow a normal distribution (p <= 0.05)")
```

# Q 23) Calculate the t scores of 95% confidence interval, 96% confidence interval, 99% confidence interval for sample size of 25

```python
import scipy.stats as stats
```

```python
sample_size = 25
```

```python
confidence_level_95 = 0.95
```

```python
t_score_95 = stats.t.ppf(1 - (1 - confidence_level_95) / 2, df=sample_size - 1)
```

```python
print("t-score for 95% confidence interval:", t_score_95)
```

```python
confidence_level_96 = 0.96
```

```python
t_score_96 = stats.t.ppf(1 - (1 - confidence_level_96) / 2, df=sample_size - 1)
```

```python
print("t-score for 96% confidence interval:", t_score_96)
```

```python
confidence_level_99 = 0.99
```

```python
t_score_99 = stats.t.ppf(1 - (1 - confidence_level_99) / 2, df=sample_size - 1)
```

```python
print("t-score for 99% confidence interval:", t_score_99)
```
